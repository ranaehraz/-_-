# Data Preprocessing

```python
# loading train data
df2 = pd.read_csv(r'C:\Users\HP\Documents\Data Analysis\Portfolio Projects\Machine learning\Projects\Titanic Classification\titanic classification test.csv')
df2.head()
```

```python
df2.info()
```

```python
df = pd.concat([df1, df2], axis = 0)
df = df.reset_index()
df
```

```python
#filling missing values using mean for numerical variables

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
```

```python
df.isnull().sum()
```

```python
#dropping columns

df = df.drop(columns='Cabin', axis = 1)
```

```python
#filling missing values using mode for categorical variable

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# #data normalization using log funct.

df['Fare'] = np.log(df['Fare']+1)
sns.distplot(df['Fare'])
```

```python
# df  = df.drop(columns='index', axis=1)
df = df.drop(columns=['Name', 'Ticket'], axis = 1)
```

```python
df
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
cols = ['Sex','Embarked']
le = LabelEncoder()
for col in cols:
    df[col] = le.fit_transform(df[col])
df.head()
```
